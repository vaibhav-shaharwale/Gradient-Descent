```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=150, n_features=1, noise=50, random_state=4)   #Generate Random sample data
x = np.reshape(x, x.shape[0])
plt.scatter(x ,y)
```

```python
# Gradient Descent function
def gradient_descent(x,y):
    m = 0   # Slope
    b = 0   # Intercept
    n = float(len(x))   # Sample Size of training set
    alpha = 0.1     # learning Rate
    epochs = 50   # Number of iterations
    prev_cost = 0   # Variable to be compared with cost value 

    # Gradient Descent Algorithm
    for i in range(epochs):
        y_pred = m*x + b    # current predicted value of y
        cost = (1/n) * sum([val**2 for val in (y_pred-y)])    # MSE function used as the Cost function
        Dm = (2/n) * sum(x*(y_pred-y))     # partial derivative wrt m
        Db = (2/n) * sum(y_pred-y)     # partial derivative wrt b
        m = m - alpha * Dm      # update current value of m
        b = b - alpha * Db      # update current value of b
        
        # compare previous cost with current cost, stop loop when cost are similar
        #if math.isclose(cost, prev_cost):
            #print('Gradient Descent reached its minimum value
            #\n Iterations : {}  | cost: {}'.format(i, round(cost,3)))
            #break
        
        prev_cost = cost    # Update Previous cost value
        print("Iteration: {} | m = {} | b ={} | error = {}".format(i, m, b, cost))

    return m, b

# Call the function
m, b = gradient_descent(x=x, y=y)

# print m & b optimal value and the Hypothesis
print('Slope (m):={} | Y-intercept (b):={}'.format(round(m,3), round(b, 3)))
print('y = {}x + {}'.format(round(m,2), round(b, 2)))
```

```python
y_pred = m*x+b
plt.figure(figsize=(8,4), dpi=100)
plt.title("Linear Regression using Gradient Descent Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis", rotation = 0)
plt.text(0.4, -90, "y = {}x +({})".format(round(m,2), round(b,2)), color='red', fontsize=10)
plt.scatter(x, y, color='#3176f7', label='trainig data', edgecolors='black', s=60, alpha=0.7)
plt.plot(x, y_pred, color='red', label='linear regression', linewidth=2.5)
plt.grid(color='blue', linestyle='--', linewidth=1, alpha=0.2)
plt.legend(loc='upper left')
plt.show()
```
